**60- Write a Python program that finds all pairs of elements in a list whose sum is equal to a given value.

```python
l1 = []
size = int(input("enter the size of the numbers: "))
for i in range(size):
  l1.append(int(input("enter the numbers: ")))
def p(l1):
  x = (int(input("the number: ")))
  newl = []
  for i in l1 :
    for j in l1:
      if i+j == x:
        newl.append((i,j))
      else:
        continue
  return  newl
print(p(l1))
```
